# 스케쥴러가 무엇이고, 단기/중기/장기로 나누는 기준


>시스템 내에는 여러 개의 프로세스가 존재하는데 자원을 할당하는 프로세스(시간과 공간)을 선택해야하는데 이 역할을 맡은 것을 스케줄러라고 합니다.

<br>

>스케줄러의 목적은 시스템 성능 향상이며 대표적인 시스템 성능 지표로응답 시간, 작업 처리량, 자원 활용도가 있는데 목적에 따라 다양한 성능 지표를 고려하여 스케줄러를 선택한다.

<br>

## 스케쥴러는 발생하는 빈도와 할당하는 자원

**장기**

>장기 스케줄러(Long-term Scheduler)는 시스템에 제출할 자가업을 결정하는 Job Scheduling(Jon->created)에서 사용하고, 시스템 내에 프로세스 수를 조절하고 중요한 것은 CPU든 I/O든 모든 써서 효율성을 높이기 위해서 I/O bounded와 compute-bounded 프로세스들을 잘 섞어서 선택해야 한다.

<br>

**중기**

>중기 스케줄러(Mid-term Scheduler)는 메모리 할당을 결정하는 Memory Allocation(suspended ready → ready)에서 사용한다.


<br>

**단기**

>단기 스케줄러(Short-term Scheduler)는 메모리 할당 받을 프로세스를 결정하는 Process Scheduling(ready->running)에서 사용하며, 가장 빈번하게 발생하므로 매우 빨라야 한다.

<br>

### 응답시간 vs 작업 처리량 vs 자원 활용도

- 응답시간(response time): 작업 요청으로부터 응답을 받을 때까지의 시간

- 작업 처리량(throughput): 단위 시간 동안 완료된 작업의 수

- 자원 활용도(resource time): 주어진 시간동안 자원이 활용된 시간

<br>

### I/O Bounded 프로세스 vs Compute-bounded 프로세스

- I/O Bounded 프로세스: I/O 대기시간이 긴 프로세스

- Compute-bounded 프로세스: CPU 사용시간이 긴 프로세스
